```python
import requests
from bs4 import BeautifulSoup
```

```python
req=requests.get("https://www.espncricinfo.com/records/tournament/batting-most-runs-career/indian-premier-league-2023-15129")
```

```python
soup=BeautifulSoup(req.content,"html.parser")
```

```python
print(soup.title.get_text())
```

```python
table=soup.find('table')
```

```python
table_data=[]
if table:
    for row in table.find_all('tr'):
        row_data=[]
        cells=row.find_all(['th','td'])
        for cell in cells:
            row_data.append(cell.text.strip())
        table_data.append(row_data)
else:
    print('Table not found')
```

```python
import pandas as pd
```

```python
if table_data:
    df=pd.DataFrame(table_data[1:],columns=table_data[0])
    print(df)
else:
    print('None')
```

```python
#Let's split player and franchise
df[['Player','Franchise']]=df['Player'].str.split(' \(', expand=True)
```

```python
df.head(3)
```

```python
#Let's remove bracket from the franchise
df['Franchise']=df['Franchise'].str.replace(')','')
```

```python
df.head(3)
```

```python
df['NO']=df['NO'].str.replace('-','0')
df['HS']=df['HS'].str.replace('-','0')
df['HS']=df['HS'].str.replace('*','0')
df['Ave']=df['Ave'].str.replace('-','0')
df['100']=df['100'].str.replace('-','0')
df['50']=df['50'].str.replace('-','0')
df['0']=df['0'].str.replace('-','0')
df['4s']=df['4s'].str.replace('-','0')
df['6s']=df['6s'].str.replace('-','0')
```

```python
df.info()
```

```python
#Changing the dtypes of the columns
df['Mat']=df['Mat'].astype(int)
df['Inns']=df['Inns'].astype(int)
df['NO']=df['NO'].astype(int)
df['Runs']=df['Runs'].astype(int)
df['HS']=df['HS'].astype(int)
df['Ave']=df['Ave'].astype(float)
df['BF']=df['BF'].astype(int)
df['SR']=df['SR'].astype(float)
df['100']=df['100'].astype(int)
df['50']=df['50'].astype(int)
df['0']=df['0'].astype(int)
df['4s']=df['4s'].astype(int)
df['6s']=df['6s'].astype(int)
```

```python
#Filtering Players with good Average+Strike rate
Filtered=df[(df['Ave']>40) & (df['SR']>160)]
```

```python
#Lets create a new column balls faced 
df['Balls Faced']=df['Runs']/(df['SR']/100)
```

```python
df['Balls Faced']=df['Balls Faced'].astype(int)
```

```python
#Ssorting Players with highest strike rate (minimum 10 Inns or 100+ balls faced)
HighSR=df.sort_values(by='SR',ascending=False)
```
